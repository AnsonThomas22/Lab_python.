```python
import math

#7
def get_triangle_sides():
    side1 = float(input("Enter the first side of the triangle: "))
    side2 = float(input("Enter the second side of the triangle: "))
    side3 = float(input("Enter the third side of the triangle: "))
    return side1, side2, side3

def calculate_area(side1, side2, side3):
    s = (side1 + side2 + side3) / 2
    area = math.sqrt(s * (s - side1) * (s - side2) * (s - side3))
    return area

print("Enter the sides for the first triangle:")
side1, side2, side3 = get_triangle_sides()
area1 = calculate_area(side1, side2, side3)


print("Enter the sides for the second triangle:")
side4, side5, side6 = get_triangle_sides()
area2 = calculate_area(side4, side5, side6)

print("Enter the sides for the second triangle:")
side7, side8, side9 = get_triangle_sides()
area3 = calculate_area(side7, side8, side9)


total_area = area1 + area2 + area3
Percent1 = (area1 / total_area) * 100
Percent2 = (area2 / total_area) * 100
Percent3 = (area3 / total_area) * 100


print(f"The area of the first triangle is: {area1:.2f}")
print(f"The area of the second triangle is: {area2:.2f}")
print(f"The area of the second triangle is: {area3:.2f}")
print(f"The total area enclosed by both triangles is: {total_area:.2f}")
print(f"The first triangle contributes: {Percent1:.2f}%")
print(f"The second triangle contributes: {Percent2:.2f}%")
print(f"The third triangle contributes: {Percent3:.2f}%")
```
